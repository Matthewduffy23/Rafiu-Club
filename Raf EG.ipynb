```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from tabulate import tabulate

# --- CONFIG ---
included_leagues = [ 'England 1.', 'Spain 1.', 'England 2.', 'Czech 1.', 'France 1.', 'Germany 1.',
                   'Belgium 1.', 'Italy 1.',]  # Only include these leagues

features = [
    'Defensive duels per 90', 'Aerial duels per 90', 'Aerial duels won, %', 'PAdj Interceptions',
    'Non-penalty goals per 90', 'xG per 90', 'Shots per 90', 'Shots on target, %', 'Goal conversion, %',
    'Crosses per 90', 'Accurate crosses, %', 'Dribbles per 90', 'Successful dribbles, %',
    'Offensive duels per 90', 'Touches in box per 90', 'Progressive runs per 90', 'Accelerations per 90',
    'Passes per 90', 'Accurate passes, %', 'xA per 90', 'Smart passes per 90', 'Key passes per 90',
    'Passes to final third per 90', 'Passes to penalty area per 90', 'Accurate passes to penalty area, %',
    'Deep completions per 90'
]

weight_factors = {
    'Passes per 90': 2, 
    'Accurate passes, %': 2, 
    'Dribbles per 90': 2,
    'Non-Penalty goals per 90': 2,
    'Shots per 90': 2, 
    'Successful dribbles, %': 2,
    'Aerial duels won, %': 2, 
    'xA per 90': 2, 
    'xG per 90': 2,
    'Touches in box per 90': 2,
}

league_strengths = {
        'England 1.': 100.00, 'Italy 1.': 97.14, 'Spain 1.': 94.29, 'Germany 1.': 94.29, 'France 1.': 91.43,
    'Brazil 1.': 82.86, 'England 2.': 71.43, 'Portugal 1.': 71.43, 'Argentina 1.': 71.43,
    'Belgium 1.': 68.57, 'Mexico 1.': 68.57, 'Turkey 1.': 65.71, 'Germany 2.': 65.71, 'Spain 2.': 65.71,
    'France 2.': 65.71, 'USA 1.': 65.71, 'Russia 1.': 65.71, 'Colombia 1.': 62.86, 'Netherlands 1.': 62.86,
    'Austria 1.': 62.86, 'Switzerland 1.': 62.86, 'Denmark 1.': 62.86, 'Croatia 1.': 62.86,
    'Japan 1.': 62.86, 'Korea 1.': 62.86, 'Italy 2.': 62.86, 'Czech 1.': 57.14, 'Norway 1.': 57.14,
    'Poland 1.': 57.14, 'Romania 1.': 57.14, 'Israel 1.': 57.14, 'Algeria 1.': 57.14, 'Paraguay 1.': 57.14,
    'Saudi 1.': 57.14, 'Uruguay 1.': 57.14, 'Morocco 1.': 57.00, 'Brazil 2.': 56.00, 'Ukraine 1.': 54.29,
    'Ecuador 1.': 54.29, 'Spain 3.': 54.29, 'Scotland 1.': 54.29, 'Chile 1.': 51.43, 'Cyprus 1.': 51.43,
    'Portugal 2.': 51.43, 'Slovakia 1.': 51.43, 'Australia 1.': 51.43, 'Hungary 1.': 51.43,
    'Egypt 1.': 51.43, 'England 3.': 51.43, 'France 3.': 48.00, 'Japan 2.': 48.00, 'Bulgaria 1.': 48.57,
    'Slovenia 1.': 48.57, 'Venezuela 1.': 48.00, 'Germany 3.': 45.71, 'Albania 1.': 44.00, 'Serbia 1.': 42.86,
    'Belgium 2.': 42.86, 'Bosnia 1.': 42.86, 'Kosovo 1.': 42.86, 'Nigeria 1.': 42.86, 'Azerbaijan 1.': 50.00,
    'Bolivia 1.': 50.00, 'Costa Rica 1.': 50.00, 'South Africa 1.': 50.00, 'UAE 1.': 50.00, 'Georgia 1.': 40.00,
    'Finland 1.': 40.00, 'Italy 3.': 40.00, 'Peru 1.': 40.00, 'Tunisia 1.': 40.00, 'USA 2.': 40.00,
    'Armenia 1.': 40.00, 'North Macedonia 1.': 40.00, 'Qatar 1.': 40.00, 'Uzbekistan 1.': 42.00,
    'Norway 2.': 42.00, 'Kazakhstan 1.': 42.00, 'Poland 2.': 38.00, 'Denmark 2.': 37.00, 'Czech 2.': 37.14,
    'Israel 2.': 37.14, 'Netherlands 2.': 37.14, 'Switzerland 2.': 37.14, 'Iceland 1.': 34.29,
    'Ireland 1.': 34.29, 'Sweden 2.': 34.29, 'Germany 4.': 34.29, 'Malta 1.': 30.00, 'Turkey 2.': 35,
    'Canada 1.': 28.57, 'England 4.': 28.57, 'Scotland 2.': 28.57, 'Moldova 1.': 28.57, 'Austria 2.': 25.71,
    'Lithuania 1.': 25.71, 'Brazil 3.': 25.00, 'England 7.': 25.00, 'Slovenia 2.': 22.00, 'Latvia 1.': 22.86,
    'Serbia 2.': 20.00, 'Slovakia 2.': 20.00, 'England 9.': 20.00, 'England 8.': 15.00, 'Montenegro 1.': 14.29,
    'Wales 1.': 12.00, 'Portugal 3.': 11.43, 'Northern Ireland 1.': 11.43, 'England 5.': 11.43,
    'Andorra 1.': 10.00, 'Estonia 1.': 8.57, 'England 10.': 5.00, 'Scotland 3.': 0.00, 'England 6.': 0.00  # Only these are relevant now
}

percentile_weight = 0.7
actual_value_weight = 0.3
league_weight = 0.3
market_value_weight = 0.3

# --- ANALYSIS ---
target_player = 'R. Durosinmi'

# Step 1: Filter Data
df_filtered = df[df['Position'].str.startswith(('CF'))].dropna(subset=features)
df_filtered = df_filtered[df_filtered['League'].isin(included_leagues)]
df_filtered['Market value'] = pd.to_numeric(df_filtered['Market value'], errors='coerce')

# Get target player row
target_row = df_filtered[df_filtered['Player'] == target_player].iloc[0]
target_vector = target_row[features].values

# Optional: Override market value
manual_override = 7000000  # Set to number like 16_000_000 if needed

# Final market value resolution
if manual_override is not None:
    target_market_value = manual_override
elif pd.isna(target_row['Market value']) or target_row['Market value'] == 0:
    target_market_value = 2_000_000  # Default fallback if missing or 0
else:
    target_market_value = target_row['Market value']

# Step 2: Club Profiles
club_profiles = df_filtered.groupby(['Team', 'Position'])[features].mean().reset_index()
scaler = StandardScaler()
scaled_club_features = scaler.fit_transform(club_profiles[features])
target_scaled = scaler.transform([target_vector])[0]

# Step 3: Similarity Calculation
weights = np.array([weight_factors.get(f, 1) for f in features])
distance = np.linalg.norm((scaled_club_features - target_scaled) * weights, axis=1)
club_profiles['Distance'] = distance
club_profiles['Club Fit %'] = ((1 - (distance - distance.min()) / (distance.max() - distance.min())) * 100).round(2)

# Step 4: League Strength Adjustments
team_league_map = df_filtered.groupby('Team')['League'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else x.iloc[0])
club_profiles['League'] = club_profiles['Team'].map(team_league_map)
club_profiles['League Strength'] = club_profiles['League'].map(league_strengths).fillna(0)
target_league_strength = league_strengths.get(target_row['League'], 1)

# League filters
club_profiles = club_profiles[
    (club_profiles['League Strength'] >= 66) &
    (club_profiles['League Strength'] <= 101)
]

diff_ratio = (club_profiles['League Strength'] / target_league_strength).clip(0.5, 1.2)
club_profiles['Adjusted Fit %'] = (
    club_profiles['Club Fit %'] * (1 - league_weight) +
    club_profiles['Club Fit %'] * diff_ratio * league_weight
)

league_gap = (club_profiles['League Strength'] - target_league_strength).clip(lower=0)
league_penalty = (1 - (league_gap / 100)).clip(lower=0.7)
club_profiles['Adjusted Fit %'] *= league_penalty

# Step 5: Market Value Fit
team_market_values = df_filtered.groupby('Team')['Market value'].mean()
club_profiles['Avg Team Market Value'] = club_profiles['Team'].map(team_market_values)

# Drop teams with missing market value data
club_profiles = club_profiles.dropna(subset=['Avg Team Market Value'])

# Calculate market value similarity score
value_fit_ratio = (club_profiles['Avg Team Market Value'] / target_market_value).clip(0.5, 1.5)
value_fit_score = (1 - abs(1 - value_fit_ratio)) * 100

# Combine into final fit score
club_profiles['Final Fit %'] = (
    club_profiles['Adjusted Fit %'] * (1 - market_value_weight) +
    value_fit_score * market_value_weight
)

# Output Top Results
top_fits = club_profiles.sort_values('Final Fit %', ascending=False).head(15).reset_index(drop=True)

# Display output table
print("\nðŸ” Top Club Fits for Target Player (Filtered to Included Leagues):\n")
print(tabulate(top_fits[['Team', 'Position', 'League', 'Final Fit %']], headers='keys', tablefmt='fancy_grid', showindex=range(1, 16)))

# Show Target Player Info
print(f"\nðŸ“Œ Target Player: {target_row['Player']}")
print(f"Team: {target_row['Team']}")
print(f"Position: {target_row['Position']}")
print(f"League: {target_row['League']}")
print(f"Market Value Used: â‚¬{target_market_value:,.0f}")
```
